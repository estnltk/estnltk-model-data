```python
import pandas as pd
import re
```

### Original manually labeled sentences

```python
free = pd.read_csv(".../obl_free.csv", sep=';', encoding='utf-8')
free
```

```python
bound = pd.read_csv(".../obl_bound.csv", sep=';', encoding='utf-8')
bound
```

### Sample 500 sentences from bound and free phrases

```python
free_samp = free.sample(n=500, random_state=1).drop_duplicates(subset='short_sent')
free_samp
```

```python
bound_samp = bound.sample(n=500, random_state=1).drop_duplicates(subset='short_sent')
bound_samp
```

```python
samples = pd.concat([free_samp, bound_samp])
samples
```

```python
samples.loc[samples['type'] == 'bound', 'removetype'] = "no"
samples.loc[samples['type'] == 'free', 'removetype'] = "yes"
```

```python
samples
```

### Additional column where short_sent is mostly untokenized

```python
short_sents = list(samples["short_sent"])
```

```python
untoks = []
for sent in short_sents:
    sent2 = sent.replace(" .", ".").replace(" ,", ",").replace(" !", "!").replace(" ?", "?").replace(" %", "%")
    sent3 = sent2.replace(" )", ")").replace("( ", "(").replace(" :", ":").replace(" %", "%")
    untoks.append(sent3)
```

```python
samples["short_sent_untoken"] = untoks
```

```python
samples
```

```python
samples.to_csv("obl_gpt_input_large1.csv", index=False, sep=";", encoding='utf-8')
```
